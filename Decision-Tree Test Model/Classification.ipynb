# Import Libraries and Load Dataset

```python
import idx2numpy
import numpy as np

# Load the MNIST Train dataset
train_images = idx2numpy.convert_from_file('train-images.idx3-ubyte')
train_labels = idx2numpy.convert_from_file('train-labels.idx1-ubyte')

# Load the MNIST Test dataset
test_images = idx2numpy.convert_from_file('t10k-images.idx3-ubyte')
test_labels = idx2numpy.convert_from_file('t10k-labels.idx1-ubyte')

#Verify the shapes of the datasets
print("Train Images Shape:", train_images.shape)
print("Train Labels Shape:", train_labels.shape)
print("Test Images Shape:", test_images.shape)
print("Test Labels Shape:", test_labels.shape)
```

# Prepare Data For Decision Tree Model
Note: Decision Trees in scikit-learn expect 2D input: so every single 28 × 28 image is being converted into a single long row (vector) containing 784 pixel values.

ie Train Images Shape: (60000, 28, 28) becomes (60000, 784)

```python
# Flatten the images for use in a classifier
X_train = train_images.reshape(train_images.shape[0], -1)
X_test = test_images.reshape(test_images.shape[0], -1)

y_train = train_labels
y_test = test_labels

# Verify the shapes of the flattened datasets
print("Flattened Train Images Shape:", X_train.shape)
print("Flattened Test Images Shape:", X_test.shape)
```

# Decision Tree Classification Model from Scikit-learn

```python
from sklearn.tree import DecisionTreeClassifier

# Create a Decision Tree Classifier
dt_classifier = DecisionTreeClassifier(random_state=42)

# Fit the model to train / learn data
dt_classifier.fit(X_train, y_train)

# Stores the predicted values for the dataset in the variable y_pred for accuracy evaluation later on
y_pred = dt_classifier.predict(X_test)
```

# Check Accuracy For Models

```python
from sklearn.metrics import accuracy_score

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred) # compares the true labels (y_test) with the predicted labels (y_pred)

print(f"Accuracy of the Decision Tree Classifier: {accuracy:.2f}")

# Test for overfitting by evaluating the model on both the training and testing datasets
print("Training Accuracy:", dt_classifier.score(X_train, y_train))
print("Testing Accuracy:", dt_classifier.score(X_test, y_test))
```
