# Import the data 
Goal is to Also try to to find the prediction column in the dataset

```python
import pandas as pd

df = pd.read_csv('hw6.data.csv', header=None) # since there are no column names in the dataset, we set header=None to prevent pandas from treating the first row as column names

print("The columns in the dataset are:")
print(df.columns)

print("Number of rows and columns in the dataset:")
print(df.shape)

print("The first 5 rows of the dataset:")
print(df.head())
```

# Split the data into 70% Training, 30% Test

```python
# library for splitting the dataset into training and testing sets
from sklearn.model_selection import train_test_split

# Split the dataset into features (X) and target variable (y) where X is what the models learn & y is what the models predict

X = df.drop(10, axis=1) # removes the target column and drop it 
y = df[10] # Extracts the target column that was dropped--- what we want to predict


# Split the dataset into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Print the shapes of the resulting training and testing sets
print(X_train.shape)
print(X_test.shape)
```

# Pick A  Classification Model from Scikit-learn

```python
from sklearn.tree import DecisionTreeClassifier

# Create a Decision Tree Classifier
dt_classifier = DecisionTreeClassifier(random_state=42)

# Fit the model to train / learn data
dt_classifier.fit(X_train, y_train)

# Stores the predicted values for tt dataset in the variable y_pred for accuracy evaluation later on
y_pred = dt_classifier.predict(X_test)
```

# Check Accuracy For Models

```python
from sklearn.metrics import accuracy_score

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred) # compares the true labels (y_test) with the predicted labels (y_pred)

print(f"Accuracy of the Decision Tree Classifier: {accuracy:.2f}")

# Test for overfitting by evaluating the model on both the training and testing datasets
print("Training Accuracy:", dt_classifier.score(X_train, y_train))
print("Testing Accuracy:", dt_classifier.score(X_test, y_test))
```
